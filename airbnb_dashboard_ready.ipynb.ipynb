```python
import pandas as pd
import numpy as np
```

```python
df_chicago = pd.read_csv("listings_chicago.csv.gz.csv")
df_neworleans = pd.read_csv("listings_neworleans.csv.gz.csv")
```

```python
df_chicago['city'] = 'Chicago'
df_neworleans['city'] = 'New Orleans'
```

```python
df= pd.concat([df_chicago , df_neworleans], ignore_index = True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df[['id','city','neighbourhood','latitude','longitude',
         'property_type','room_type','price',
         'number_of_reviews','reviews_per_month',
         'host_id','host_name','availability_365',
         'calculated_host_listings_count']]
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['neighbourhood'].fillna('Unknown',inplace = True)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['price'],inplace = True)
```

```python
df.dropna(subset=['host_id'],inplace = True)
df.dropna(subset=['host_name'], inplace=True)
```

```python
df['reviews_per_month'].fillna(0, inplace=True)
```

```python
df.isnull().sum()
#“I removed irrelevant columns like description and URLs because they were not required for my analysis and would not add value to the dashboard.”
```

```python
# Reason : “I handled missing values carefully by dropping critical missing values like price, while imputing non-critical ones like neighbourhood to avoid data loss.”
```

```python
df['price'].head()
df.dtypes
```

```python
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.dtypes
```

```python
df['price_category'] = pd.cut(
    df['price'],
    bins=[0,100,300,1000,60000],
    labels=['Low','Medium','High','Luxury']
)
```
